**Modifying Keys in a JSON File**

```python
import json

# Load the original JSON file (replace with your actual path)
json_input_path = "./data/your_input_file.json"  # Replace with your actual path
with open(json_input_path, "r", encoding="utf-8") as json_file:
    json_data = json.load(json_file)

# Modify keys in each JSON object
# Example: replace key "old_key_name" with "new_key_name"
for item in json_data:
    if isinstance(item, dict):
      item['instruction']= 'Find from the following input whether it is toxic or safe.'
      item['input']= item.pop('sentence',None)
      item['output']= item.pop('abusive',None)
      del item['class_vote']
      del item['abusive_vote']
      del item['abusive_class']


def replace_output(data):
    if isinstance(data, dict):
        for key, value in data.items():
            if key == 'output':
                if value == 'no':
                    data[key] = 'safe'
                elif value == 'yes':
                    data[key] = 'toxic'
            else:
              replace_output(value)
    elif isinstance(data, list):
        for item in data:
            replace_output(item)

replace_output(data)

extracted_comments = []
for post in data['ptt_post']:
    for comment in post['comment']:
        extracted_comments.append(comment)

for entry in data:
    if 'output' in entry:
        entry['output'] = str(entry['output'])
json_string = json.dumps(data, indent=4, ensure_ascii=False)
json_string = json_string.replace('": ', '": ').replace(',\n    "input"', ',\n    "input"')


# Save the modified JSON data to a new file
json_output_path = "./data/your_modified_file.json"  # Replace with your actual path
with open(json_output_path, "w", encoding="utf-8") as json_file:
    json.dump(json_data, json_file, ensure_ascii=False, indent=4)
```
